# Table1

```python
# Create a table with appropriate column names, and use double quotes for column names with spaces
cursor.execute('''
CREATE TABLE table1 (SNo INTEGER, Hotelname VARCHAR, City VARCHAR, Location VARCHAR,
Numberofreviews INTEGER,
Userratings INTEGER,
Starrating INTEGER,
Reviewscore VARCHAR
)''')

# Read data from a CSV file using pandas
data_tab = pd.read_csv('E:\Surendra Data\DATA Scientist Course Detail\Internship\Capstone Project\Data\Data1.csv')  # Replace with the path to your CSV file

# Insert data into the SQLite table
data_tab.to_sql('table1', db, if_exists='append', index=False)
```

1) Write a query to give details of hotels city is Sydney and ratings is more than 5.

2) Write a query to find the location with the most reviews.

```python
# query to find the location with the most reviews.
results = cursor.execute("SELECT Location, MAX(Numberofreviews) AS Max_Reviews FROM table1")
for i in results:
    print(i)
```

2) Write a SQL query to calculate the average price for each room type.

5) Write a query to calculate the average price for each room type (Use table2 and Table 3) 

```python
results = cursor.execute("SELECT table3.'RoomType', AVG(table3.Prices) AS 'Average Price' FROM Table2 AS table2 JOIN Table3 AS table3 ON table2.'sno' = table3.'sno' GROUP BY table3.'RoomType';")
for i in results:
    print(i)
```
